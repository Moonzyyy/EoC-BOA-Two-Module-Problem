```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
reimplement_results = json.load(open("reimplement.txt"))
boa_results = json.load(open("BOA.txt"))
```

```python
reimplement_values = sorted(reimplement_results.keys())
reimplement_means = [float(reimplement_results[n]['mean']) for n in reimplement_values]
reimplement_stds = [float(reimplement_results[n]['std']) for n in reimplement_values]
for i in range(len(reimplement_values)):
        reimplement_values[i] = int(reimplement_values[i])
        
        
boa_values = sorted(boa_results.keys())
boa_means = [float(boa_results[n]['mean']) for n in boa_values]
boa_stds = [float(boa_results[n]['std']) for n in boa_values]
for i in range(len(boa_values)):
        boa_values[i] = int(boa_values[i])


plt.figure(figsize=(10, 6)) 
plt.plot(boa_values, boa_means, 
             marker='o', linestyle='-', color='red', linewidth=2, 
             markersize=6, label='BOA')

for i, (x, y, std) in enumerate(zip(boa_values, boa_means, boa_stds)):
        plt.plot([x, x], [y-std, y+std], color='red', linewidth=1)
        plt.plot([x-0.5, x+0.5], [y-std, y-std], color='red', linewidth=1)
        plt.plot([x-0.5, x+0.5], [y+std, y+std], color='red', linewidth=1)

plt.plot(reimplement_values, reimplement_means, 
             marker='s', linestyle='-', color='blue', linewidth=2, 
             markersize=6, label='reimplement results')

for i, (x, y, std) in enumerate(zip(reimplement_values, reimplement_means, reimplement_stds)):
        plt.plot([x, x], [y-std, y+std], color='blue', linewidth=1)
        plt.plot([x-0.5, x+0.5], [y-std, y-std], color='blue', linewidth=1)
        plt.plot([x-0.5, x+0.5], [y+std, y+std], color='blue', linewidth=1)

plt.xlabel('n', fontsize=12)
plt.ylabel('generation to peak', fontsize=12)
plt.legend(loc='upper left', fontsize=11)
plt.grid(True, alpha=0.3)
plt.xlim(0, 90)
plt.ylim(0, 2000)
plt.xticks(range(0, 91, 10))
plt.yticks(range(0, 2001, 200))

plt.tight_layout()
plt.show()
```
